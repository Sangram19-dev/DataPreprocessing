```python
data.info()
```

```python
#Outlier Treatment

#rainfall and n_hot_rooms are outlier so we will do treatment of these 2 var

np.percentile(data.n_hot_rooms,[99])
```

```python
uv = np.percentile(data.n_hot_rooms,[99])[0] #uv=uppervalue
```

```python
data[(data.n_hot_rooms>uv)] #get the values who have more than max percentage(n_hot_rooms)
```

```python
data.n_hot_rooms[(data.n_hot_rooms>3*uv)]= 3*uv #adding queries changing values
```

```python
data[(data.n_hot_rooms>uv)] #modified
```

```python
#for rainfall

lv= np.percentile(data.rainfall,[1])[0] #here using lower value
```

```python
data[(data.rainfall<lv)]
```

```python
data.rainfall[(data.rainfall<0.3*lv)]= 0.3*lv
```

```python
data[(data.rainfall<lv)]
```
